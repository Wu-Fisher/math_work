```python
import math
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
```

定义龙格库塔函数

```python
def RuKu(f,a,b,init,N):
    x = np.linspace(a,b,N+1)
    y = np.zeros(N+1)
    y[0] = init
    h = (b-a)/N
    xc = x[0]
    yc = y[0]
    print("打印结果保留四位小数：")
    print(f"x0 = {xc:.4f}, y0 = {yc:.4f}")
    for i in range(1,N+1):
        K1 = h*f.evalf(subs={'x':xc,'y':yc})
        K2 = h*f.evalf(subs={'x':xc+h/2,'y':yc+K1/2})
        K3 = h*f.evalf(subs={'x':xc+h/2,'y':yc+K2/2})
        K4 = h*f.evalf(subs={'x':xc+h,'y':yc+K3})
        yn = yc + (K1+2*K2+2*K3+K4)/6
        y[i] = yn
        print(f"x{i} = {round(x[i],4)},y{i} = {round(y[i],4)}")
        xc = x[i]
        yc = y[i]
        
    return x,y
```

定义检测精度函数

```python
def Verf(X,Y,exp):
    Y2 = [exp.evalf(subs={'x':x}) for x in X]

    E=0
    
    for i in range(len(X)):
        E += (Y[i]-Y2[i])**2
    E = math.sqrt(E/len(X))
    print(f"验证得到标准差为 = {E}")

    print("作图分析误差：")
    plt.scatter(X,Y,label="RUBU")
    plt.plot(X,Y2,'r',label='REAL')
    plt.legend()
    plt.show()

    return E
```

问题3（1）
$$
\frac{dy}{dx}=-20\left( y-x^2 \right) ,0\le x\le 1,N=5,10,20
\\
y\left( 0 \right) =\frac{1}{3}
\\
real: y=x^2+\frac{1}{3}e^{-20x}
$$


```python
x , y =sp.symbols('x,y')
f =-20*(y-x**2)+2*x
a = 0
b = 1
init = 1/3
exp = x**2 +(1/3)*sp.exp(-20*x)
for n in [5,10,20]:
    print("N = ",n)
    X,Y = RuKu(f,a,b,init,n)
    E = Verf(X,Y,exp)
    

```

问题3（2）

$$
\frac{dy}{dx}=-20y+20\sin x+\cos x,0\le x\le 1,N=5,10,20
\\
y\left( 0 \right) =1
\\
real: y=e^{-20x}+\sin x
$$


```python
x , y =sp.symbols('x,y')
f =-20*y + 20*sp.sin(x) + sp.cos(x)

a = 0
b = 1
init = 1
exp =sp.exp(-20*x) + sp.sin(x)
for n in [5,10,20]:
    print("N = ",n)
    X,Y = RuKu(f,a,b,init,n)
    E = Verf(X,Y,exp)
    
```

问题3（3）

$$
\frac{dy}{dx}=-20\left( y-e^x\sin x \right) +e^x\left( \sin x+\cos x \right) ,0\le x\le 1,N=5,10,20
\\
y\left( 0 \right) =0
\\
real: y=e^x\sin x
$$


```python
x , y =sp.symbols('x,y')
f =-20*(y-sp.exp(x)*sp.sin(x))+sp.exp(x)*(sp.sin(x)+sp.cos(x))

a = 0
b = 1
init = 0
exp =sp.exp(x)*sp.sin(x)
for n in [5,10,20]:
    print("N = ",n)
    X,Y = RuKu(f,a,b,init,n)
    E = Verf(X,Y,exp)
```
