```python
import numpy as np
import sympy as sp
import math
```

```python
def minTWO(x,y)->float:
    n=len(x)
    a=0
    for i in range(n):
       a=a+x[i]*y[i]

    return a 
```

```python
x=sp.symbols('x')
exp1=x-x+1
exp2=x**2
h=[19,25,31,38,44]
a = [exp1.evalf(subs={x:i}) for i in h]
b = [exp2.evalf(subs={x:i}) for i in h]
c = [19,32.3,49,73.3,97.8]

A = [[minTWO(a,a),minTWO(a,b)],[minTWO(b,a),minTWO(b,b)]]
B = [[minTWO(c,a)],[minTWO(c,b)]]
print(A)
print(B)
```

```python
A=np.array(A,dtype=float)
B=np.array(B,dtype=float)
ans=np.linalg.solve(A,B)
print(ans)
```

```python
exp= ans[0][0]*exp1+ans[1][0]*exp2
print(sp.simplify(exp))
```

```python
d=0
for  i in range(len(h)):
    d=d+(c[i]-exp.evalf(subs={x:h[i]}))**2
print(math.sqrt(d))
```
